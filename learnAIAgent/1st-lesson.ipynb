# TOOLS 

```python
def calc_tool(query):
    return str(eval(query))
```

```python
def simple_agent(user_input):

    decision_prompt = f"""You are a helpful AI agent assistant. 
    You will be given a user input and you need to decide if you 
    should use the calculator tool or respond directly.

    If calculator tool is needed, respond with "use calculator: <calculation query>"
    
    If calculator tool is not needed, respond with "respond directly: <your response>"
    User input: {user_input}
    """

    decision_response = llm2.invoke(decision_prompt).content.strip()
    if decision_response.startswith("use calculator:"):
        calc_query = decision_response.replace("use calculator:", "").strip()
        calc_result = calc_tool(calc_query)
        return f"Calculator result: {calc_result}"
    elif decision_response.startswith("respond directly:"):
        direct_response = decision_response.replace("respond directly:", "").strip()
        return f"Direct response: {direct_response}"
    else:
        return "Invalid response from agent."
```

```python
print (simple_agent("what is weather"))
```

```python
print (simple_agent("12 * 12"))
```

```python
print (simple_agent("calculate 45 * 23"))
```
